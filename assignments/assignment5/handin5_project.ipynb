## Import

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colormaps
```

## Read data and create year, above and below arrays

```python
readArray = np.genfromtxt("Land_and_Ocean_summary.txt", dtype=None, comments="%")
year = np.array([ elem[0] for elem in readArray ])
above = np.array([ elem[1] for elem in readArray ])
below = np.array([ elem[5] for elem in readArray ])
```

## Plot 1

```python
fig, allyears = plt.subplots()  
allyears.plot(year, above, color="r", label="Temperature Anomaly (°C) above Ice")
allyears.plot(year, below, color="b", label="Temperature Anomaly (°C) below Ice")
plt.xlabel("Year")
plt.ylabel("°C")
plt.title("Global Below and Above Ice relative Mean")
plt.legend(["First line", "Second line"])

plt.show() 
```

## Plot 2

```python
ten_year_indices = np.where(np.mod(year, 10) == 0)
ten_year_x = year[ten_year_indices]
ten_year_above = above[ten_year_indices]
ten_year_below = below[ten_year_indices]

fig, tenyears = plt.subplots()  
tenyears.plot(ten_year_x, ten_year_above, color="r", label="Temperature Anomaly (°C) above Ice")
tenyears.plot(ten_year_x, ten_year_below, color="b", label="Temperature Anomaly (°C) below Ice")
plt.xlabel("Year")
plt.ylabel("°C")
plt.title("Global Below and Above Ice: Decade Mean 1850-2018")
plt.legend(["First line", "Second line"])

plt.show() 
```

## Plot 3

```python
my_cmap = colormaps["coolwarm"]

fig, color_bars = plt.subplots()
for i in range(len(year)):
    plt.bar(year[i], above[i], my_cmap(i))
plt.xlabel("Year")
plt.ylabel("°C")
plt.title("Global Surface Temperature relative Mean")
plt.legend(["First line"])

plt.show()
```
